```python
def hypothesis(a,b,c,d,x1,x2,x3):
    return a*x1+b*x2+c*x3+d
```

```python
def error(a,b,c,d,x1,x2,x3,y):
    err=0
    m=len(x1)
    
    for i in range(m):
        err+=np.power(hypothesis(a,b,c,d,x1[i],x2[i],x3[i])-y[i],2)
        
    return (1/(2*m))*err
```

```python
def step_gradient(a,b,c,d,x1,x2,x3,y,learning_rate):
    grad_a=0
    grad_b=0
    grad_c=0
    grad_d=0
    
    m=len(x1)
    
    for i in range(m):
        grad_a=(1/m)*(hypothesis(a,b,c,d,x1[i],x2[i],x3[i])-y[i])*x1[i]
        grad_b=(1/m)*(hypothesis(a,b,c,d,x1[i],x2[i],x3[i])-y[i])*x2[i]
        grad_c=(1/m)*(hypothesis(a,b,c,d,x1[i],x2[i],x3[i])-y[i])*x3[i]
        grad_d=(1/m)*(hypothesis(a,b,c,d,x1[i],x2[i],x3[i])-y[i])
        
    a=a-(grad_a*learning_rate)
    b=b-(grad_b*learning_rate)
    c=c-(grad_c*learning_rate)
    d=d-(grad_d*learning_rate)
    
    return a,b,c,d
```

```python
def descend(init_a,x1,init_b,x2,init_c,x3,init_d,y,learning_rate,iterations):
    a=init_a
    b=init_b
    c=init_c
    d=init_d
    
    for i in range(iterations):
        e=error(a,b,c,d,x1,x2,x3,y)
        
        if i%1000 == 0:
            print(f"Error:{np.sqrt(e)},a:{a},b:{b},c:{c},d:{d}")
            
        
        a,b,c,d=step_gradient(a,b,c,d,x1,x2,x3,y,learning_rate)
    
    return a,b,c,d
```
