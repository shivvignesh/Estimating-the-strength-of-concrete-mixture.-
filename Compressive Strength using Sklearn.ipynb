```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_squared_error,r2_score

from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_excel('Concrete_data.xls')
data.head()
```

```python
x=np.array(data['Cement (component 1)(kg in a m^3 mixture)'])
y=np.array(data['Concrete compressive strength(MPa, megapascals) '])
```

```python
scaler=MinMaxScaler()
```

```python
x=x.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
x=scaler.fit_transform(x)
y=scaler.fit_transform(y)
```

```python
y
```

```python
x
```

```python
plt.plot(x,y,"r.")
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
y_pred
```

```python
for i in zip(y_pred,y_test):
    print(i)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
plt.scatter(x_train,y_train,color="red")
plt.plot(x_train,regressor.predict(x_train),"b")

plt.xlabel('Cement (component 1)(kg in a m^3 mixture)')
plt.ylabel('Concrete compressive strength(MPa, megapascals) ')
plt.title("Linear Regression(train)")
```

```python
plt.scatter(x_test,y_test,color="blue")
plt.plot(x_train,regressor.predict(x_train),"r")

plt.xlabel('Cement (component 1)(kg in a m^3 mixture)')
plt.ylabel('Concrete compressive strength(MPa, megapascals) ')
plt.title("Linear Regression(test)")
```
