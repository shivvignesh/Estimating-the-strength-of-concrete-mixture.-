```python
X=np.array(x)
Y=np.array(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3)
```

```python
x_test.size
```

```python
def hypothesis(a,b,x):
    return a*x+b
```

```python
def error(a,b,x,y):
    err=0
    m=len(x)
    
    for i in range(m):
        err+=np.power(hypothesis(a,b,x[i])-y[i],2)
        
    return (1/(2*m))*err
```

```python
def step_gradient(a,b,x,y,learning_rate):
    grad_a=0
    grad_b=0
    m=len(x)
    
    for i in range(m):
        grad_a=(1/m)*(hypothesis(a,b,x[i])-y[i])*x[i]
        grad_b=(1/m)*(hypothesis(a,b,x[i])-y[i])
        
    a=a-(grad_a*learning_rate)
    b=b-(grad_b*learning_rate)
    
    return a,b
```

```python
def descend(init_a,x,init_b,y,learning_rate,iterations):
    a=init_a
    b=init_b
    
    for i in range(iterations):
        e=error(a,b,x,y)
        
        if i%1000 == 0:
            print(f"Error:{np.sqrt(e)},a:{a},b:{b}")
            
        
        a,b=step_gradient(a,b,x,y,learning_rate)
    
    return a,b 
```
